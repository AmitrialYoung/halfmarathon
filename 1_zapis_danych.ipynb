```python
load_dotenv()
```

### Inicjalizacja klienta do komunikacji z chmurą

```python
s3 = boto3.client('s3',)
```

```python
# nazwa bucketu możliwa do kontrolowania dzieki zmiennej.

BUCKET_NAME = "half-marathon"
```

### Zapis danych na Digital Ocean Spaces

```python
# Ścieżka do lokalnego katalogu
local_directory = 'dane'

# Przechodzimy przez wszystkie pliki i podfoldery w katalogu 'dane'
for root, dirs, files in os.walk(local_directory):
    for filename in files:
        # Tworzymy pełną ścieżkę lokalną do pliku
        local_path = os.path.join(root, filename)
        
        # Tworzymy ścieżkę docelową w Digital Ocean Spaces (zachowując strukturę)
        # relpath sprawi, że plik 'dane/2023/plik.csv' trafi do 'dane/2023/plik.csv' w chmurze
        relative_path = os.path.relpath(local_path, os.path.dirname(local_directory))
        #Zamiana \ na / dla chmury
        s3_path = relative_path.replace("\\", "/")
        
        print(f"Przesyłanie: {local_path} -> {s3_path}...")
        
        # Wysyłka pliku
        s3.upload_file(local_path, BUCKET_NAME, s3_path)

print("--- Zakończono przesyłanie całego katalogu ---")
```
